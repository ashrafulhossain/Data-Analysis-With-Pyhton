### how many people have heard disease and how many people doesn't have heart disease?

```python
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind = 'bar', color=['orchid', 'salmon'], figsize= (5, 4))
plt.title('Heart Disease Values')
plt.xlabel('1 = Disease, 0 = No Disease')
plt.ylabel('Total Peoples')
plt.show()
plt.savefig('Heart Disease Values.png', format='png', dpi=300, bbox_inches='tight')
```

### people of which sex has most heart disease ?

```python
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie', figsize=(5,4))
plt.legend(['Male', 'Female'])
plt.show()
```

```python
pd.crosstab(df.sex, df.target)
```

```python
sns.countplot(x='target', data=df, hue='sex')
plt.savefig('Target.png', format='png', dpi=300, bbox_inches='tight')
```

###  Age vs Maximum heart rate when people have heart disease
###  Age vs Maximum heart rate when people have heart disease

```python
sns.displot( x = 'age', data = df, bins = 30, kde = True);
plt.savefig('Age Distribution.png', format='png', dpi=300, bbox_inches='tight')
```

```python
sns.displot(x = 'thalach', data = df, bins = 30, kde = True, color = 'chocolate');
plt.savefig('Thalach Distribution.png', format='png', dpi=300, bbox_inches='tight')
```

```python
# Creating a figure
plt.figure(figsize=(8,6))

#plotting the values for people who have heart disease
plt.scatter(df.age[df.target==1], 
            df.thalach[df.target==1], 
            c="tomato") # color

#plotting the values for people who doesn't have heart disease
plt.scatter(df.age[df.target==0], 
            df.thalach[df.target==0], 
            c="lightgreen") # color

# Addind info
plt.title("Heart Disease w.r.t Age and Max Heart Rate")
plt.xlabel("Age")
plt.legend(["Disease", "No Disease"])
plt.ylabel("Max Heart Rate");
plt.savefig('Heart Disease w.r.t Age and Max Heart Rate.png', format='png', dpi=300, bbox_inches='tight')
```

## Age vs serum cholestoral when people have heart disease.
## Age vs serum cholestoral when people doesn't have heart disease.

```python
# Creating another figure
plt.figure(figsize=(8,6))

#plotting the values for people who have heart disease
plt.scatter(df.age[df.target==1], 
            df.chol[df.target==1], 
            c="salmon") # define it as a scatter figure

#plotting the values for people who doesn't have heart disease
plt.scatter(df.age[df.target==0], 
            df.chol[df.target==0], 
            c="lightblue") # axis always come as (x, y)

# Add some helpful info
plt.title("Heart Disease w.r.t Age and Serum Cholestoral")
plt.xlabel("Age")
plt.legend(["Disease", "No Disease"])
plt.ylabel("Serum cholestoral");
plt.savefig('Heart Disease w.r.t Age and Serum Cholestoral.png', format='png', dpi=300, bbox_inches='tight')
```
